```python
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
print(f"x_train array : {x_train}")
print(f"y_train array : {y_train}")
```

```python
# m - Number of training examples
m = x_train.shape[0] # m = len(x_train)
print(f"m : {m}")
```

```python
for i in range(m):
    print(f"(x^({i}),y^({i})) = ({x_train[i]},{y_train[i]})")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("Housing prices")
plt.ylabel("Prices in 1000 dollars")
plt.xlabel("Prices in 1000 sq.feet")
plt.show()
```

```python
w = 200
b = 100
```

```python
def compute_model_output(w,b,m,x=np.array([1.2,1.4])):
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
temp_f_wb = compute_model_output(w,b,m)
x=np.array([1.2,1.4])
plt.scatter(x,temp_f_wb,label="Predicted Output")
plt.scatter(x_train,y_train,marker="x",c="r", label="Actual Training examples")
plt.title("Housing prices")
plt.ylabel("Prices in 1000 dollars")
plt.xlabel("Prices in 1000 sq.feet")
plt.legend()
plt.show()
```
